PANDAS
It has two data structure which are Series and Dataframe.
###### SERIES

```python
import pandas as pd
import numpy as np
```

```python
#STRING SERIES
fruits = pd.Series(["apples", "oranges", "bananas"])

print("MY FRUIT SERIES")
print(fruits)
```

```python
#FLOAT SERIES
temperature = pd.Series([32.6, 34.1, 28.0, 35.9], index = ["a","b","c","d"])

print("TEMPERATURE IN CELSIUS")
print(temperature, "\n")
```

```python
#INTEGER SERIES
factors_of_12 = pd.Series([1,2,4,6,12], name = "factors of 12")

print("FACTORS OF 12 SERIES")
print(factors_of_12, "\n")
```

```python
#USING INDEX
print ("2nd fruit: ", fruits.iloc[1])
```

The difference between loc and iloc is that loc works labelling that is, accessing via keyword while iloc is based on integer positioning

```python
print ("2nd fruit: ", fruits[1], "\n")

#USING KEY
print ("temperature at key \"b\": ", temperature.loc['b'])
```

```python
print ("temperature at key \"b\": ", temperature.iloc[1])
```

DATAFRAME

```python
import pandas as pd

fruits_jack = ["apples", "oranges", "bananas"]
fruits_john = ["guavas", "kiwis", "strawberries"]
index = ["a", "b", "c"]
all_fruits = {"Jack's": fruits_jack, "John's": fruits_john}

fruits = pd.DataFrame(all_fruits, index = index)
print(fruits, "\n")
```

```python
new_fruits = fruits.reset_index(drop = True)
print(new_fruits, "\n")
# to change the indexing used
```

```python
#USING INDEX
print("1st fruit:")
print(fruits.iloc[0], "\n")
```

```python
#USING KEY
print("Fruits at key \"c\":")
print(fruits.loc["c"])
```

```python
#USING COLUMN NAME
print("Jack's fruits: ")
print(fruits["Jack's"], "\n")
```

```python
#CHAINED QUERY
print("Johns third fruit: ")
print(new_fruits["John's"][2], "\n")
```
