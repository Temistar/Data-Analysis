Let create a Pandas Dataframe from a dictionary having keys and values. The dictionary contains list of subjects and how many times a particular student passed each subject.

#### To slice or select data in a dataframe is done in two major ways using; loc or iloc

loc[[row]],

loc[[row], [column]],

loc[[:], [column]]

```python
df.head()
```

```python
df.loc[['s5']]
```

```python
df.loc[['s5'], ['title']]
```

```python
df.loc[['s5', 's7', 's8'], ['type','title', 'country']]
```

```python
df.loc[:, ['title', 'type', 'director']]
```

#### For iloc data is selected using integers

```python
df.iloc[[1, 4]]
```

```python
df.iloc[[1,4, 6],[1,3,5]]
```
